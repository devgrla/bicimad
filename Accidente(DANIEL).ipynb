## ACCIDENTES POR DISTRITOS



```python

import pandas as pd
ac_2018 = pd.read_csv('dat/AccidentesBicicletas_2018.csv',encoding='ISO-8859-1',sep=';')

ac_2 = ac_2018.apply(lambda an: an.str.strip()if an.dtype=='object'else an)

tmp= ac_2.groupby('DISTRITO')['* Nº VICTIMAS'].sum()
tmp.sort_values(ascending=False)
```

## ¿QUE DIA HAY MAS VICTIMAS?


```python
ac_2 = ac_2018.apply(lambda an: an.str.strip()if an.dtype=='object'else an)

tmp= ac_2.groupby('DIA SEMANA')['* Nº VICTIMAS'].sum()
tmp.sort_values(ascending=False)
```
